# Image matching algorithm using opencv

The algorithm used in the code is ORB (Oriented FAST and Rotated BRIEF), which combines FAST for keypoint detection and BRIEF for descriptor generation, with added rotation invariance. It is used in this code for detecting and matching keypoints between two images.
